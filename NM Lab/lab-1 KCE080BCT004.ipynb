```python
plt.figure(figsize = (10,6)) # Set the figure size
x = np.linspace(-5,5,100) # 100 points between -5 and 5
plt.plot(x, x**2, "ko", label = "quadratic") # plot the quadratic function
plt.plot(x, x**3, "r*", label = "cubic") # plot the cubic function
plt.title(f"Plot of Various Polynomials from {x[0]} to {x[-1]} ") # gives title to the figure
plt.xlabel("X axis") # labels the x axis 
plt.ylabel ("Y axis") # labels the y axis
plt.legend(loc = "lower right") # adds a legend to the plot
plt.xlim(-6.6)  
plt.ylim(-10,10)
plt.grid() # turns on or off the grid lines
plt.show () 
```

**Given the lists x = np.arange(11) and y = x2, create a 2 x 3 subplot where each subplot plots x
versus y using plot, scatter, bar, loglog, semilogx, and semilogy. Title and label each plot
appropriately. Use a grid, but a legend is not necessary here.**

```python

x = np.arange (11)
y= x**2
plt.figure(figsize = (14, 8))
plt.subplot(2, 3, 1)
plt.plot (x,y)
plt.title ("Plot")
plt.xlabel ("X")
plt.ylabel ("Y")
plt.grid()
plt.subplot(2, 3, 2)
plt.scatter (x,y)
plt.title ("Scatter")
plt.xlabel ("X")
plt.ylabel ("Y")
plt.grid()
plt.subplot(2, 3, 3)
plt.bar (x,y)
plt.title ("Bar")
plt.xlabel ("X")
plt.ylabel ("Y")
plt.grid()
plt.subplot(2, 3, 4)
plt.loglog (x, y)
plt.title ("Loglog")
plt.xlabel ("X")
plt.ylabel("Y")  
plt.grid(which="both")
plt.subplot (2, 3, 5)
plt.semilogx (x, y)
plt.title ("Semilogx")
plt.xlabel ("X")
plt.ylabel ("Y")
plt.grid(which="both")
plt.subplot (2, 3, 6)
plt.semilogy (x,y)
plt.title ("Semilogy")
plt.xlabel ("X")
plt.grid()
plt.tight_layout ()
plt.show () 


```

```python
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt

plt.style.use("seaborn-v0_8-poster")
fig = plt.figure(figsize = (10, 10))

ax = plt.axes(projection = "3d")

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X,Y = np.meshgrid(x,y)
Z = np.sin(np.sqrt(X**2 + Y**2))

ax.plot_surface(X, Y, Z, cmap='viridis')

plt.show()
```
